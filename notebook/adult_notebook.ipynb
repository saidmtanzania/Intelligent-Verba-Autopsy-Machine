```python
data = pd.read_csv('../data/ihme_adult_data.csv')

print(f'Data Shape: {data.shape}')
```

```python
data.newid = data.apply(lambda x:  f'<a href="/{x.birth_year}">{x.last_age}</a>', axis=1)

data
```

```python
data = data[data.possible_COD != 'Other Non-communicable Diseases']
data = data[data.possible_COD != 'Other Cardiovascular Diseases']
data = data[data.possible_COD != 'Other Infectious Diseases']
data = data[data.possible_COD != 'AIDS']


data
```

```python
main_cols=[  
            'birth_year', 'deceased_sex', 'death_year', 'last_age', 'marital_status', 'last_education', 'education_years', 'respondent_sex',
            'respondent_relationship', 'had_asthma', 'had_arthritis', 'had_cancer', 'had_copd', 'had_dementia', 'had_depression', 'had_deabetes',
            'had_epilepsy', 'had_heart_disease', 'had_hypertension', 'had_obesity', 'had_stroke', 'had_tb', 'had_aids', 'care_sought',
            'care_from_traditional_healer', 'care_from_govt_hosp', 'care_from_private_hospital', 'care_from_private_physician',
            'care_from_phrmacy', 'have_health_records', 'show_health_records', 'death_certificate_issued', 'show_death_certificate', 
            'used_cigarattes', 'drank_alcohol', 'alcohol_amount', 'memory_loss_experience_3_last_months','experience_loss_of_conciousness', 
            'had_headaches', 'vomit_final_week', 'had_cough', 'cough_blood', 'breathing_difficulty', 'fast_breathing',
            'had_sores', 'had_fever', 'fever_severity', 'sweat_wit_fever', 'had_rash', 
            ]
```

```python
data.possible_COD.value_counts().sort_values(ascending=False)
```

```python
X=data[main_cols]
y=data.possible_COD

X.shape, y.shape
```

```python
marital_status_map={'Never Married':0, 'Married':1, 'Divorced':2, 'Separated':3, 'Widowed':4, 'Unknown':999}
last_education_map={'High School':2, 'Primary School':1, 'No Schooling':0, 'College or Higher':3, 'Unknown':999}
x_yes_no_map={'Yes':1, 'No':0, "Don't Know":9, 'Refused to Answer':8}
fever_severity_map = {"Don't Know":9, 'Moderate':2, 'Mild':1, 'Severe':3}
alcohol_amount_map = {'High':3, "Don't Know":9, 'Moderate':2, 'Low':1}
respondent_relationship_map = {"Mother":1, "Father":2, "Grandmother":3, "Grandfather":4, "Aunt":5, "Uncle":6, "Husband":7,"Wife":8,
          "Brother":9, "Sister":10, "Birth Attendant":11, "Other Male":12, "Other Female":13, "Unknown":999}
yes_no_map={'Yes':1, 'No':0, "Don't Know":9}
deceased_sex_map={'Male':0, 'Female':1}

dont_know_map={"Don't Know":999}
```

```python
X.marital_status=X.marital_status.replace(marital_status_map)
X.last_education=X.last_education.replace(last_education_map)
X.deceased_sex=X.deceased_sex.replace(deceased_sex_map)
X.fever_severity=X.fever_severity.replace(fever_severity_map)
X.alcohol_amount=X.alcohol_amount.replace(alcohol_amount_map)
X.respondent_relationship=X.respondent_relationship.replace(respondent_relationship_map)
X.respondent_sex=X.respondent_sex.replace(deceased_sex_map)

for col in X.columns.tolist():
    vals=X[col].unique().tolist()

    if('Refused to Answer' in vals):
        X[col]=X[col].replace(x_yes_no_map)
    elif('Yes' in vals):
        X[col]=X[col].replace(yes_no_map)
    elif("Don't Know" in vals):
        X[col]=X[col].replace(dont_know_map)

# X.head()
```

```python
for col in X.columns:
    X[col]=X[col].fillna(X[col].mode()[0])
    X[col]=X[col].astype('int64')

# X.isna().sum().sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import f1_score, accuracy_score

x_train, x_test, y_train, y_test=tts(X, y, test_size=0.2, stratify=y, random_state=568)

x_train.shape, x_test.shape
```

```python
from lightgbm import LGBMClassifier

model=LGBMClassifier(
                        n_estimators=50,
                        colsample_bytree=0.6,
                        learning_rate=0.09,
)


model.fit(x_train, y_train)
```

```python
#initialize the algorithm
from sklearn.model_selection import StratifiedKFold

f1_scores=[]

skf=StratifiedKFold(n_splits=5, random_state=568)

for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    #fit the model
    model.fit(X_train, y_train)

    #make predictions for evaluations
    y_preds=model.predict(X_test)

    #display score
    print('F1 Score: ', f1_score(y_test, y_preds, average='weighted'))
    f1_scores.append((f1_score(y_test, y_preds, average='weighted')))

print(f'\nMean F1 Score: {np.mean(f1_scores)}')
```
